```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pickle
import argparse
%matplotlib inline 
```

```python
image = mpimg.imread('../img/signs_vehicles_xygrad.jpg')

plt.imshow(image)
```

```python
hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
plt.imshow(hls)
# ax2.set_title('HLS', fontsize=50)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(hsv)
```

```python
H = hls[:,:,0]
L = hls[:,:,1]
S = hls[:,:,2]
```

```python
f, axarr = plt.subplots(2, 2)

thresh = (90, 255)
binary = np.zeros_like(S)
binary[(S > thresh[0]) & (S <= thresh[1])] = 1
axarr[0,1].imshow(binary)

binary[(H > thresh[0]) & (H <= thresh[1])] = 1
axarr[1,0].imshow(binary)

binary[(L > thresh[0]) & (L <= thresh[1])] = 1
axarr[1,1].imshow(binary)


axarr[0,0].imshow(hls)
plt.show()
```

```python
# now first convert to gray

f, axarr = plt.subplots(2, 2)

thresh = (90, 255)
binary = np.zeros_like(S)
binary[(S > thresh[0]) & (S <= thresh[1])] = 1
axarr[0,1].imshow(binary)

binary[(H > thresh[0]) & (H <= thresh[1])] = 1
axarr[1,0].imshow(binary)

binary[(L > thresh[0]) & (L <= thresh[1])] = 1
axarr[1,1].imshow(binary)


axarr[0,0].imshow(hls)
plt.show()
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(2, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=50)

hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
ax2.imshow(hls)
ax2.set_title('HLS', fontsize=50)

hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
ax3.imshow(hsv)
# ax3.set_title('HSV', fontsize=50)
#plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
plt.show()
```
